```python
#mainly to be implemented if the data is in the form of clusters and are not corellated
#the tree cassifer can also be used for the same thing
import pandas as pd
import numpy as np
import sklearn
from sklearn import preprocessing
from sklearn.model_selection import cross_validate
from sklearn import neighbors
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv('cancer.csv')
df
```

```python
df['lable']=df['no.1']
df['norecurrenceevents']=df['no-recurrence-events']

df=df.drop(['no-recurrence-events'],1)
df=df.drop(['no.1'],1)
df=df.drop(['left_low'],1)

df
```

```python
df.lable[df.lable == 'yes'] = 1
df.lable[df.lable == 'no'] = 0

df.norecurrenceevents[df.norecurrenceevents == 'no-recurrence-events'] = 0
df.norecurrenceevents[df.norecurrenceevents == 'recurrence-events'] = 1

df.left[df.left == 'right']= 0
df.left[df.left == 'left']= 1

df.no[df.no=='no']=0
df.no[df.no=='yes']=1

df.premeno[df.premeno == 'ge40']= -1
df.premeno[df.premeno == 'lt40']= 0
df.premeno[df.premeno == 'premeno']= 1

#correcting the range
```
