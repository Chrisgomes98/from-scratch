```python
x=[1,2,3,4,5,6]
y=[2,6,5,6,8,10]

plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def slope(x,y):
    for i in range(len(x)):
        k1=[0]*len(x)
        k2=[0]*len(x)
        k1[i]=x[i]*y[i]
        k2[i]=x[i]*x[i]
    m=((mean(x)*mean(y))-mean(k1))/((mean(x)*mean(x))-mean(k2))
    return m

m=slope(x,y)
print(m)    
```

```python
def y_int(x,y):
    b=mean(y)-m*mean(x)
    return b
b=y_int(x,y)
print(b)
```

```python

#y=mx+b
Y=[0]*len(x)
X=[0]*len(x)
for i in range(len(x)):
    X[i]=x[i]
    Y[i]=m*x[i]+b
```

```python
plt.scatter(x,y)
plt.plot(X,Y) #this is the regression line where X=x Y=m*x+b
plt.show()
```

```python
def predict(p):
    q=m*p+b
    return q
    
```
